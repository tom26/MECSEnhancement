## Figure 2. Surface Renewal Theories.

For absorption accompanied by irreversible first-order reaction, expressions for $J/(k_Lc^*)$ (where $J$ is the gas flux, $c^*$ the physical gas concentration at the surface, and $k_L$ the liquid-side mass transfer coefficient) are derived in Danckwerts' classic *Gas-Liquid Reactions*, p.105-109. Note that due to the linearity of the differential equations and the surface renewal theories, the extension to reversible reactions simply requires $c^*$ be replaced by $\Delta c = c^*-c_{eq}$, where $c_{eq}$ is the concentration of gas at equilibrium with the current bulk liquid composition.


```julia
using PyPlot
M = logspace(-3,3,10000)
R1 = sqrt(M)./(tanh(sqrt(M)))
R2 = π./(4*sqrt(M)).*((4M/π+1/2).*erf(2*sqrt(M/π)) + 2*sqrt(M)/π.*exp(-4M/π))
R3 = sqrt(1+M)

PyPlot.loglog(M,R1,":k",linewidth=1.0)
PyPlot.loglog(M,R2,"--k",linewidth=1.0)
PyPlot.loglog(M,R3,"-k",linewidth=1.0)
PyPlot.legend(["Whitman", "Higbie", "Danckwerts"], fontsize=15.0)
PyPlot.xlabel(L"$Ha^2$",fontsize=15.0)
PyPlot.ylabel(L"$J/(k_LΔc)$",fontsize=15.0);
```

## Figure 3. Particle Size and Thiele Modulus
This is simply a plot of $(\phi \coth \phi - 1)/(\phi^2/3)$, which quantifies the influence of particle size on the absorption rate. 

```julia
ϕFigure3=linspace(0.00001,10,1000)
PyPlot.plot(ϕFigure3,(ϕFigure3.*coth(ϕFigure3)-1)./(ϕFigure3.^2/3),"k")
PyPlot.xlim([0,10])
PyPlot.ylim([0.2,1])
PyPlot.xlabel(L"ϕ",fontsize=15)
PyPlot.ylabel(L"\frac{\phi \coth\phi - 1}{\phi^2/3}",fontsize=18);
```

## Figures 4-7. Mass Transfer Correlations

We first create functions for each of the mass transfer correlations. We then use these functions to create several plots. 

### Onda et al 1968
$$
k_L=0.0051(a_td_p)^{0.4}\left(\frac{\mu g}{\rho}\right)^{1/3}\left(\frac{L}{a_w\mu}\right)^{2/3} \left(\frac{\mu}{\rho \mathcal{D}}\right)^{-1/2}
$$

where $g$ is the gravitational constant, $d_p$ is an equivalent packing diameter, defined by $d_p = 6(1-\varepsilon)/a_t$ at the bottom of p58 of Onda's paper, $\varepsilon$ is the voidage in the (dry) column, $L$ is the liquid flow rate per unit area of column, $\mathcal{D}$ is the diffusivity of the gas in the liquid, $\rho$ and $\mu$ are the liquid density and viscosity respectively, and $a_w$ and $a_t$ are the wetted and total packing surface area respectively. These are in turn related by

$$
\frac{a_w}{a_t} = 1 - \exp\left[-1.45\left(\frac{\sigma_c}{\sigma}\right)^{0.75}\left(\frac{L}{a_t \mu}\right)^{0.1}\left(\frac{L^2 a_t}{\rho^2 g}\right)^{-0.05}\left(\frac{L^2}{\rho \sigma a_t}\right)^{0.2} \right]
$$

where $\sigma_c$ is the critical surface tension of the packing material, and $\sigma$ is the surface tension of the solvent


```julia
function Onda_kL(D,d_p,L,μ,ρ,a_t,σ_c,σ)
    g = 9.81                                                                              #Gravitational Constant, m/s2
    k_L = 0.0051 * (a_t*d_p)^0.4 * (μ*g/ρ)^(1/3) * (L/(a_w*μ))^(2/3) * (μ/(ρ*D))^(-1/2)                          
    a_w = a_t * (1-exp(-1.45 * (σ_c/σ)^0.75 * (L/(a_t*μ))^0.1 * (L^2*a_t/(ρ^2*g))^-0.05 * (L^2/(ρ*σ*a_t))^0.2))
    return [k_L, a_w]
end;
```

### Shulman et al 1955

$$
\frac{k_L d_s}{\mathcal{D}} =25.1 \left(\frac{d_sL}{\mu}\right)^{0.45}\left(\frac{\mu}{\rho\mathcal{D}}\right)^{0.5}
$$

where $d_s$ is the diameter of a sphere with the same surface area as the packing, $L$ is the liquid flow rate per unit area of column, $\mathcal{D}$ is the diffusivity of the gas in the liquid, and $\rho$ and $\mu$ are the liquid density and viscosity respectively. The effective surface area was presented graphically by Shulman et al (1955) and was fitted by Treyball (1980) to the following correlation:

$$
a = d\left[\frac{808G}{\sqrt{\rho_G}}\right]^q L^h
$$

where $G$ is the gas flux through the column and $\rho_G$ is the gas density (note that this equation is dimensional, and will only be accurate in SI units.) According to Treybal, this correlation can be extrapolated up to about $10 kg/m^2.s$

```julia
function Shulmann_kL(D,d_s,L,μ,ρ,G,ρ_gas)
    #Values of Constants d, q and h for 1/2-inch Raschig Rings. c.f. Treybal (1980), p.205.
    d = L < 2.0 ? 28.01 : 14.69
    q = L < 2.0 ? 0.2323*L-0.30 : 0.01114*L+0.148
    h = L < 2.0 ? -1.04 : -0.11
    
    k_L = 25.1 * (D/d_s) * (d_s*L/μ)^0.45 * (μ/(ρ*D))^0.5
    a = d * (808*G/(ρ_gas^0.5))^q * L^h
    return [k_L, a]
end;
```

### Billet and Schultes (1993)

$$
k_L = C_L \left(\frac{\rho g}{\mu}\right)^{1/6}\left(\frac{\mathcal{D}}{d_h}\right)^{1/2}\left(\frac{L}{\rho a_t}\right)^{1/3}
$$

where $C_L$ is a packing dependent constant and $d_h$ is the hydraulic diameter of the packing, defined by $4\varepsilon/a_t$. They combine this with a correlation for mass transfer area. For liquids whose surface tension doesn't decrease down the column:

$$
\frac{a}{a_t} = 1.5(a_td_h)^{-0.5}\left(\frac{Ld_h}{\mu}\right)^{-0.2}\left(\frac{L^2d_h}{\sigma\rho}\right)^{0.75}\left(\frac{L^2}{gd_h\rho^2}\right)^{-0.45}
$$

For liquids whose surface tension decreases, the account for the Marangoni effect by multiplying this result by $1-2.3\times10^{-4}\big|Ma\big|^{0.5})$ where Ma is the Marangoni number, defined as:

$$
Ma = \frac{d\sigma}{dx}\frac{\Delta x}{\mathcal{D}\mu a}
$$


```julia
function BandS_kL(D,d_h,L,μ,ρ,a_t,σ,C_L)
    g = 9.81         #m/s2
    k_L = C_L * (ρ*g/μ)^(1/6) * (D/d_h)^(1/2) * (L/(ρ*a_t))^(1/3)
    a = a_t * 1.5 * (a_t*d_h)^(-0.5) * (L*d_h/μ)^-0.2 * (L^2*d_h/(σ*ρ))^0.75 * (L^2/(g*d_h*ρ^2))^-0.45
    return [k_L, a]
end;
```

### Hanley and Chen 2011

Hanley and Chen (2011) developed a number of correlations for Pall rings, metal IMTP and Mellapak and Flexipak arrangements. Their correlation could be compared with Rocha et al. and Tsai et al. for Mellapak, as well as Onda et al. and Billet and Scholes (and maybe Shulman also) for Pall rings. They all follow the same pattern, which will look familiar to followers of Shulman:

$$
\frac{k_Ld_h}{\mathcal{D}} = A\left(\frac{d_hL}{\mu}\right)^{\alpha}\left(\frac{\mu}{\rho \mathcal{D}}\right)^{\beta}
$$

here $d_h$ is the hydraulic diameter. The effective area ("liquid participating in mass transfer") is given by:

$$
\frac{a}{a_t} = C\left(\frac{\rho_V}{\rho}\right)^{\gamma}\left(\frac{\mu_V}{\mu}\right)^{\zeta}Re_L^XFr_L^{\Delta}We_L^ERe_V^{\Phi}\left(\frac{\cos\theta}{\cos\pi/4}\right)^{t}
$$

where $\theta$ is the corrugation inclination angle in radians measured from the vertical, and the exponents are various packing-dependent constants. They use the following definitions for these dimensionless groups:

$$
Re_L = \frac{d_hL}{\mu}
$$

$$
Fr_L = \frac{L^2}{\rho^2gd_h}
$$

$$
We_L = \frac{d_hL^2}{\rho\sigma}
$$

$$
Re_V = \frac{d_hG}{\mu_v}
$$

```julia
function HandC_kL(D,d_h,L,μ,ρ,a_t,σ,G,μ_v,ρ_v,costheta,Constants)
    (A,α,β,C,γ,ζ,X,Δ,E,Φ,t) = Constants
    g = 9.81      #m/s^2
    Re_L = d_h*L/μ
    Sc_L = μ/(ρ*D)
    Fr = L^2/(ρ^2*g*d_h)
    We = d_h*L^2/(ρ*σ)
    Re_V = d_h*G/μ_v
    k_L = A * (D/d_h) * (Re_L)^α * (Sc_L)^β
    a = a_t * C * (ρ_v/ρ)^γ * (μ_v/μ)^ζ * Re_L^X * Fr^Δ * We^E * Re_V^Φ * (costheta/cos(π/4))^t
    return [k_L,a]
end;
```

### Rocha et al 1996

Rocha et al. 1996 developed the following correlation, valid for many types of structured packing:

$$
k_L = 2\left(\frac{C\mathcal{D}U}{\pi S}\right)^{1/2}
$$

Here $S$ is the corrugation side length, $C$ is a constant equal to 0.9 for most structured packings, and $U$ is

$$
U = \frac{L}{\rho \varepsilon h_L \sin \alpha}
$$

where $h_L$ is the fractional liquid holdup and $\alpha$ is the structured packing corrugation angle from the horizontal. This is combined with the following correlation for effective surface area, valid for sheet-netal structured packings:

$$
\frac{a}{a_t} = F_{SE}\frac{29.12 (We\cdot Fr)^{0.15}S^{0.359}}{Re^{0.2}\varepsilon^{0.6}(1-0.93\cos \gamma) \sin^{0.3} \alpha} = F_{SE}F_t
$$

where $F_{SE}$ is a packing dependent constant and $\theta$ is the contact angle and the other parameters are dimensionless numbers defined as follows:

$$
We = \frac{L^2 S}{\rho \sigma}
$$

$$
Fr = \frac{L^2}{\rho^2 Sg}
$$

$$
Re = \frac{LS}{ \mu}
$$

$\sin \alpha$ is the same as $\cos \theta$ in Hanley and Chen's paper, and the surface tension for metal sheet packings is given by (equation 11 of Rocha et al (1996)):

$$
\cos\gamma =
\left\{
	\begin{array}{ll}
		0.90  & \mbox{if } \sigma \leq 0.055 N/m \\
		5.211 \times 10^{-16.835\sigma} & \mbox{if } \sigma \geq 0.055 N/m
	\end{array}
\right.
$$

Note that we still need to define $h_L$, the liquid holdup. This can be defined via $F_t$ in the $a/a_t$ equation above. From Rocha et al.'s first paper in the series (Rocha et al 1993):

$$
h_L = \left(4\frac{F_t}{S}\right)^{2/3}\left(\frac{3\mu L}{\rho^2 \sin \alpha \varepsilon g_{eff}}\right)^{1/3}
$$

here $g_{eff}$ is an effective gravity, but when we are far from flooding, this can safely be treated as $g$.

```julia
function Rocha_kL(D,S,L,μ,ρ,σ,sinalpha,a_t,cosgamma,ɛ,F_SE)
    g = 9.81                 #N/kg
    We = L^2*S/(ρ*σ)
    Fr = L^2/(ρ^2*S*g)
    Re = L*S/μ
    F_t = 29.12 * (We*Fr)^0.15 * S^0.359 / (Re^0.2*ɛ^0.6*(1-0.93*cosgamma)*sinalpha^0.3)
    a = a_t * F_SE * F_t
    holdupfrac = (4F_t/S)^(2/3) * (3μ*L/(ρ^2*sinalpha*ɛ*g))^(1/3)
    U = L/(ρ*ɛ*holdupfrac*sinalpha)
    k_L = 2(C*D*U/(π*S))^0.5
    return [k_L,a]
end;
```

### Tsai et al (2011)

Tsai et al (2011) gave the following function for the effective surface area for a chemically-reactive solvent:

$$
\frac{a}{a_t} = 1.34\left(\left(\frac{\rho}{\sigma}\right)g^{1/3}\left(\frac{L}{\rho}\frac{Bh}{4S}\right)^{4/3}\right)^{0.116}
$$

where $B$, $h$ and $S$ are packing specific parameters.

```julia
function Tsai_a(ρ,σ,L,a_t,B,h,S)
    g = 9.81        #m/s^2
    a = a_t * 1.34 * (ρ/σ * g^(1/3) * (L/ρ * B*h/(4S))^(4/3))^0.116
    return a
end;
```
